```python
acquired = "1980-01-01/2015-12-31"
prod_date = "1995-04-01"
ulx = -1821585
uly = 2891595
lrx = -1821085
lry = 2891095

specs_url = "http://lcmap-test.cr.usgs.gov/v1/landsat/chip-specs"
chips_url = "http://lcmap-test.cr.usgs.gov/v1/landsat/chips"
```

```python
from firebird import driver   as fd
from firebird import chip     as fc
from firebird import aardvark as fa
```

```python
band_queries = fd.chip_spec_urls(specs_url)
```

```python
band_queries
```

```python
chipids = fc.ids(ulx, uly, lrx, lry, fa.chip_specs(band_queries['blues'])[0])
```

```python
type(chipids)
```

```python
for ids in chipids:
    pyccd_inputs = fd.pyccd_rdd(specs_url, chips_url, ids[0], ids[1], acquired)
    for px in pyccd_inputs:
        print(tuple(px[0][0]))
```
